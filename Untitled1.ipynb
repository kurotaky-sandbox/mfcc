```python
wav, sr = librosa.load("hello.wav", sr=44100)
```

```python
# https://qiita.com/tom_m_m/items/91ba624dd8507bc0b746
#
import librosa.display
import matplotlib.pyplot as plt

plt.figure()
plt.figure(figsize=(15, 5))

# https://librosa.org/doc/main/generated/librosa.display.waveshow.html#librosa.display.waveshow
librosa.display.waveshow(wav, sr)
plt.show()
```

```python
import numpy as np

stft_result = librosa.stft(wav)
abs_result = np.abs(stft_result)
power_spec = librosa.amplitude_to_db(abs_result, ref=np.max)

plt.figure(figsize=(20,5))
librosa.display.specshow(power_spec, y_axis='log', x_axis='time', sr = sr)
plt.title('Power Spectrogram')
plt.colorbar(format='%+2.0f dB')

plt.tight_layout()
plt.show()
```

```python
# https://qiita.com/m__k/items/0673dedadc4c0546a58e

# 波形データとサンプルレートを取得
waveform, sample_rate = librosa.load("hello.wav", sr=44100)

# メルスペクトログラムを取得
feature_melspec = librosa.feature.melspectrogram(y=waveform, sr=sample_rate)

# MFCCを取得
feature_mfcc = librosa.feature.mfcc(y=waveform, sr=sample_rate)

# 可視化してみる
plt.figure(figsize=(20, 5))

# librosa.display.waveplotで波形データを可視化できます
plt.subplot(1,3,1)
plt.title("wave form")
librosa.display.waveplot(waveform, sr=sample_rate, color='blue')

# librosa.display.specshowでメルスペクトログラム、MFCCを可視化できます
plt.subplot(1,3,2)
plt.title("mel spectrogram")
librosa.display.specshow(feature_melspec, sr=sample_rate, x_axis='time', y_axis='hz')
plt.colorbar()

plt.subplot(1,3,3)
plt.title("MFCC")
librosa.display.specshow(feature_mfcc, sr=sample_rate, x_axis='time')
plt.colorbar()

plt.tight_layout()
plt.show()
```
